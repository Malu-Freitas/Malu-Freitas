Este primeiro grafico contém um código Python simples para criar gráficos de números ao quadrado usando Matplotlib.

```python

from matplotlib import pyplot as plt

squares = [1, 4, 9, 16, 25]

plt.plot(squares, linewidth=5)
plt.grid(True)
plt.title('Numeros Ao Quadrado', fontsize=14)
plt.xlabel('Numeros', fontsize=12)
plt.ylabel('Valores Ao Quadrado', fontsize=12)
plt.tick_params(axis= 'both', labelsize=14)
plt.show
```

Este segundo grafico tambem utiliza a biblioteca Matplotlib para criar um gráfico de linhas com marcadores ('o'). Os valores no eixo x são personalizados com rótulos 'A', 'B', 'C', 'D', 'E' e 'F'. O gráfico representa a relação entre os valores em x e y, onde x e y são listas de dados numéricos.

```python
from matplotlib import pyplot as plt

x=[1, 2, 3, 4, 5, 7]
y=[2, 1, 6, 4, 8, 5]
values = ['A', 'B', 'C', 'D', 'E', 'F']

plt.plot(x, y, marker='o')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.xticks(x, values)
plt.show()
```

O terceiro grafico, utiliza tambem a biblioteca Matplotlib para criar um gráfico de linhas que representa a média das temperaturas de dois anos (2021 e 2022) nos meses de janeiro a junho. Os dados são representados no eixo x pelos meses e no eixo y pelas temperaturas em graus Celsius. O gráfico é estilizado usando o estilo 'bmh' e inclui rótulos para os eixos e uma legenda que diferencia os anos.



```python
from matplotlib import pyplot as plt

x = ['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN']
y1 = [40, 42.5, 35, 28, 22, 20]
y2 = [38, 35, 28, 23, 24, 22]

plt.style.use('bmh')
plt.title('MEDIA DE TEMPERATURAS 1º SEM', color='gray', fontweight='bold')
plt.ylabel('GRAUS CELSIUS', fontsize=12, color='orange')
plt.xlabel('MESES', fontsize=12, color='orange')
plt.axis(ymin=0, ymax=50)

'''colocar cor na linha
plt.plot(x, y1, label='2021', color= 'green')
plt.plot(x, y2, label='2022', color= 'orange')
'''

plt.plot(x, y1, label='2021')
plt.plot(x, y2, label='2022')
plt.legend(fontsize=16, frameon=True, framealpha=0.2, facecolor='black')
plt.show
```

Este quarto grafico tambem utiliza a biblioteca Matplotlib para criar um gráfico de dispersão que representa a relação entre números e seus valores ao quadrado. Os valores são representados no eixo x como "NUMEROS" e no eixo y como "VALORES AO QUADRADO". Cada ponto no gráfico de dispersão é marcado em roxo com bordas pretas e um tamanho maior (200 unidades). O título do gráfico é "NUMEROS X QUADRADO".

```python
#grafico de dispersao
from matplotlib import pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.scatter(x, y, s=200, c='purple', edgecolors='black')
plt.title('NUMEROS X QUADRADO', fontsize=14)
plt.xlabel('NUMEROS', fontsize=10)
plt.ylabel('VALORES AO QUADRADO', fontsize=10)
plt.show()
```

Este quinto grafico é um gráfico de barras verticais que representa os valores de três produtos diferentes (Produto A, Produto B e Produto C). Cada barra vertical é proporcional ao valor associado a cada produto. O gráfico é estilizado com linhas de grade para melhor clareza.

```python
'''
GRAFICO DE BARRAS
'''
grupos = ['Produto A', 'Produto B', 'Produto C']
valores = [1, 10, 100]

plt.bar(grupos, valores) #barras verticais
plt.grid(True)
plt.show()
```

O sexto grafico, é um gráfico de barras horizontais que representa os valores de três produtos diferentes (Produto A, Produto B e Produto C). Cada barra horizontal é proporcional ao valor associado a cada produto. O gráfico é estilizado usando o estilo 'ggplot' e as barras são coloridas em azul. Além disso, os rótulos no eixo y são rotacionados em 45 graus para facilitar a leitura

```python
#grafico horizontal

grupos = ['Produto A', 'Produto B', 'Produto C']
valores = [1, 10, 100]

plt.style.use('ggplot')
plt.barh(grupos, valores, color='blue') #barras horizontais
plt.yticks(rotation=45)
plt.show()
```

O setimo grafico tambem utiliza a biblioteca Matplotlib para criar um gráfico de barras horizontais que representa a participação de mercado de cinco produtos agrícolas (Soja, Cana de Açúcar, Milho, Café e Mandioca). Cada barra horizontal representa a porcentagem de vendas de cada produto, com margens de erro indicadas por barras horizontais em tons de rosa. O gráfico é estilizado usando o estilo 'ggplot' e inclui rótulos para os eixos x e y. O tamanho da figura é personalizado para melhor clareza.

```python
produtos = ['Soja', 'Cana de Açucar', 'Milho', 'Café', 'Mandioca']
qtdepct = [33.6, 16.8, 10.3, 6.2, 3.8]
margem_erro = [3, 2, 1, 2, 2]

plt.style.use('ggplot')
plt.figure(figsize=(13, 5))
plt.title('AGRO', fontsize = 14, color = 'black')
plt.xlabel('VENDAS', fontsize=12, color = 'black')
plt.ylabel('PRODUTO', fontsize=12, color = 'black')
plt.barh(produtos, qtdepct, xerr=margem_erro, color = 'pink')
plt.show
```

O oitavo grafico tambem utiliza a biblioteca Matplotlib para criar um gráfico de barras empilhadas que representa o desempenho de três times (Time A, Time B e Time C) em três rodadas diferentes (Rodada 1, Rodada 2 e Rodada 3). Cada barra no gráfico é dividida em segmentos coloridos que representam o desempenho de cada time em cada rodada. O código utiliza o NumPy para criar um array x com três posições, controlando a posição das barras no gráfico. As barras são espaçadas uniformemente usando a largura 0.2. Os rótulos no eixo x são personalizados para indicar os times e uma legenda é adicionada para distinguir as rodadas.

```python
#COLUNAS AGRUPADAS
import numpy as np
import matplotlib.pyplot as plt

x = np. arange(3)

rodada1 = [34, 56, 12]
rodada2 = [12, 56, 78]
rodada3 = [14, 23, 45]
largura = 0.2
plt.bar(x-0.2, rodada1, largura)
plt.bar(x, rodada2, largura)
plt.bar(x+0.2, rodada3, largura)
plt.xticks(x, ['Time A', 'Time B', 'Time C'], color = 'black')
plt.legend(['Rodada 1', 'Rodada 2', 'Rodada 3'])

plt.show()
```

o nono grafico, é um gráfico de barras empilhadas que representa a distribuição de três tipos de recursos (Ouro, Pedra e Madeira) em três diferentes vilas (Vila A, Vila B e Vila C). Cada vila é representada por barras empilhadas, onde cada segmento da barra corresponde a um tipo de recurso. O código utiliza arrays NumPy para representar as quantidades de recursos em cada vila. As barras são coloridas de acordo com o tipo de recurso (Ouro em dourado, Pedra em cinza e Madeira em marrom). Uma legenda é adicionada para identificar os recursos.



```python
#COLUNAS EMPILHADAS
ouro = np.array((104, 50, 280))
pedra = np.array((233, 140, 60))
madeira = np.array((95, 228, 65))

recursos = ['Ouro', 'Pedra', 'Madeira']
vilas = ['Vila A', 'Vila B', 'Vila C']

plt.bar(vilas, ouro, color = 'gold')
plt.bar(vilas, pedra, color = 'grey', bottom = ouro)
plt.bar(vilas, madeira, color = 'brown', bottom = ouro + pedra)
plt.legend(recursos)
plt.show()
```

O decimo e ultimo grafico tambem utiliza a biblioteca Matplotlib para criar um gráfico de pizza (ou gráfico de setor) que representa a distribuição das vendas em quatro diferentes setores de negócios (E-commerce, Loja Física, E-mail e Marketplace). Cada fatia da pizza representa a proporção das vendas em um determinado setor, e o valor percentual é exibido em cada fatia usando o formato '%1.1f%%'. O código utiliza uma lista de vendas e uma lista de setores para criar o gráfico de pizza.

```python
#grafico de pizza
import matplotlib.pyplot as plt

vendas = [3000, 2300, 1000, 500]
setores = ['E-commerce', 'Loja Fisica', 'E-mail', 'Marketplace']

plt.pie(vendas, labels=setores, autopct = '%1.1f%%')

plt.show()
```
